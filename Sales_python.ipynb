```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns ={'purchase amount (usd)': 'purchase_amount'})

df.columns
```

```python
#create a column named age_groups

labels = ['Young Adult', 'Adult','Middle-Aged','Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column frequency_purchase_days

frequency_mapping = {
     'Fortnightly': 14,
     'Weekly': 7,
     'Monthly': 30,
     'Quarterly': 90,
     'Bi-Weekly':14,
     'Annually': 365,
     'Every 3 Months': 90
}
df['frequency_purchase_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','frequency_purchase_days']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head()
# check whether the two columns contain the exact same data

(df['discount_applied']==df['promo_code_used']).all()
```

```python
#delete the promo_code_used  column(axis = 1)
df = df.drop('promo_code_used', axis = 1)
```
